## DATA PREPROCESSING

```python
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
```

```python
dataset = pd.read_csv("D:/BML/Stock_Market.csv.xls")
dataset
```

# Splitting The Attributes Into Dependent And Independent:


## Independant Values

```python
X = dataset.iloc[:,:-4].values
X
```

## Dependant Values

```python
Y = dataset.iloc[:,-4:].values
Y
```

## Counting the missing values in the dataset


```python
print(dataset.isnull().sum())
```

## Dropping missing value records- remaining rows are kept as its place


```python
dataset.dropna(inplace=True)
dataset
```

```python
imputer = SimpleImputer(missing_values=np.nan,strategy='mean')
```

```python
imputer = imputer.fit(X[:,3:])
```

```python
X[:,3:] = imputer.transform(X[:,3:])
X
```

## Splitting the datatset into training and testing:

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.25,random_state=1)
```

```python
print(X_train)
print(X_test)
print(Y_train)
print(Y_test)
```

## Feature scaling:

```python
scaler = StandardScaler()
Y_train = scaler.fit_transform(Y_train)
Y_test = scaler.fit_transform(Y_test)
print(Y_train)
print(Y_test)
```
